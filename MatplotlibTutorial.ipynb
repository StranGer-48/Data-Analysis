>The **Figure** is the overall window or page that everything is drawn on. You can create multiple independent Figures. A Figure can have several other things in it, such as a suptitle, which is a centered title to the figure. You'll also find that you can add a legend and color bar, for example, to your Figure.

>To the figure you add **Axes**. The Axes is the area on which the data is plotted with functions such as **plot()** and **scatter()** and that can have ticks, labels, etc. associated with it.

```
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1,2,3,4],[10,20,50,60],color='lightblue',linewidth=3)
ax.scatter([0.3,3.8,1.2,2.5],[11,25,9,26],color='darkgreen',marker='^')
ax.set_xlim(0.5,4.5)
plt.show()
```

**Create Your Plot**


```
# Initialize a Figure
fig = plt.figure()

# Add axes to the Figure
fig.add_axes([0,0,1,1])
```

**Subplots** are used to setup and place yours Axes on a regular grid. So that means in most cases, axes and subplot are synonymous, they will designate the same thing. When you do call subplot to add Axes to your figure, do so with the **add_subplots()** function. There is, however, a difference between the **add_axes()** and the **add_subplots()** function.

```
# Create a Figure
fig = plt.figure()

# Set up Axes
ax = fig.add_subplot(111)

# Scatter the data
ax.scatter(np.linspace(0,1,5), np.linspace(0,5,5))

plt.show()
```

The difference between **fig.add_axes()** and **fig.add_subplot()** does not lie in the result: they both return an Axes object. However, they do diffe in the mechanism that is used to add the axes: you pass a list to **add_axes** which is the lower left point, the width and the height, This means that the axes object is positioned in absolute coordinates.

The **add_subplot()** function doesn't provide the option to put the axes at a certain positin: it does, however, allow the axes to be situated according to a subplot grid.

**Deleting an Axes**

If you want to remove an axes from your plot, you can use **delaxes()** to remove and update the current axes:

```
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# Plot the data
ax1.bar([1,2,3],[3,4,5])
ax2.barh([0.5,1,2.5],[0,1,2])
ax2.axhline(0.45)
ax1.axvline(0.65)
ax3.scatter(np.linspace(0, 1, 5), np.linspace(0, 5, 5))

# Delete `ax3`
fig.delaxes(ax2)

# Show the plot
plt.show()
```

**Set Plot Title and Axes Labels**

To change your plot title and axes labels, you can folloe one of the following approaches, depending of which container of which you want to make use:
  * The easiest way to set these things right is by using 
    > ax.set(title="A title", xlabel="x" ylabel="y") or
    > ax.set_xlib(), ax.set_ylim() or
    > ax.set_title()
  * If you want to work with the figure, you might also resort to **fig.suptitle()** to add a title to your plot.
  * If you are making use of the default settings that the package has to offer, you might want to use
    > plt.title(), plt.xlabel(), plt.ylabel()
  

**Fix the Plot Layout**

A thing to consider when you ar using subplots to build up your plot is the tight_layout function, which will help you to make sure that the plots fit nicely in figure. You ideally cal it after you have plotted you data and custimized your plot; So that's right before you call **plt.show()** that you should use **plt.tight_layout()**

```
# Prepare the data
x = np.linspace(0, 10, 100)
y = np.linspace(0,10,100)

# Plot the data
plt.plot(x, y, label='linear')
plt.title('Linear')
plt.xlabel('x')
plt.ylabel('y')

# Add a legend
plt.legend()

plt.tight_layout()
# Show the plot
plt.show()
```

**Customizing Matplotlib with style sheets and rcParms**

**Using style sheets**
  
  The style package adds support for easy-to-swethch plotting "styles" with the same parameters as a matplotlib rc file (whicj is read at startup to config Matplotlib).

  The are a number of pre-defined styles provided by Matplotlib. For example, there is a pre-defined style calles "ggplot", which emulates the aesthetics of ggplot(a popular plotting package for R).


```
with plt.style.context('ggplot'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```
